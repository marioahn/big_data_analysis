```python
import pandas as pd
import matplotlib.pyplot as plt

engListening = [30, 60, 90, 31, 32, 69, 92, 99]
engScore = [70, 80, 90, 70, 71, 85, 90, 92]

data = {'engListening': engListening, 'engScore': engScore}
df = pd.DataFrame(data)

plt.figure(figsize=(3,2))
plt.scatter(df['engListening'], df['engScore'])
plt.show()

coef = df.corr(method='pearson')
print(coef)
```

```python
import pandas as pd

engListening = [30, 60, 90, 31, 32, 69, 92, 99]
engReading = [40, 45, 60, 20, 15, 70, 60, 80]
engClass = [60, 120, 120, 60, 60, 180, 120, 120]
engScore = [70, 80, 90, 70, 71, 85, 90, 92]

data2 = {'engListening': engListening, 'engReading': engReading, 'engClass': engClass, 'engScore': engScore}
df2 = pd.DataFrame(data2)

# 1. 피어슨 상관분석
perasonCoef = df2.corr(method='pearson')
print(perasonCoef)
print()
# 2. 스피어만 상관분석
spearmanCoef = df2.corr(method='spearman')
print(spearmanCoef)
print()
# 3. 켄달 상관분석
kendallCoef = df2.corr(method='kendall')
print(kendallCoef)
```

```python
# 3-1. 타이타닉 데이터에서 상관관계 뽑아내기
import seaborn as sns

titanic = sns.load_dataset('titanic')


# try1
    # 두 컬럼의 상관관계 -> 단, 숫자가 아닌 male, alive여부같은 컬럼은 전처리해야 가능함
    # -> numeric으로 전처리 가능
a = titanic[['age','fare']]
two_coef = a.corr(method='pearson') 
two_coef


# try2 - > number옵션: True, False는 변환x해줌
titanic_numeric = titanic.select_dtypes(include=['number'])
titanic_numeric
all_coef = titanic_numeric.corr(method='pearson') 
all_coef


# try3: 레퍼런스 - 이게 굿. 걍 corr함수 인자에 numeric하면 되네
    # True,False값들도 컬럼 포함시켜줌(변환해서) - adult_male, alone컬럼
    # *이제 여기에서, survied열을 보면, 어떤 컬럼이 생존에 가장 도움되는 요소였는지 확인이 가능!
    # adult_male이 오히려 -0.55 -> 성인남성이 생존율이 낮았음
reference = titanic.corr(method='pearson', numeric_only=True) 
reference
```

```python
# 3-2. 타이타닉 생존율 분석2(상관분석+히트맵)
# titanic = titanic.drop('embarked_town', axis=1) # 실수로 이상한 컬럼 넣음

# 1)각 컬럼별로 null값이 몇개인지 확인하고, 아래에서 fillna작업
titanic.isnull().sum() 

titanic['age'] = titanic['age'].fillna(titanic['age'].median())
titanic['embarked'].value_counts()


titanic['deck'].value_counts()
titanic['deck'] = titanic['deck'].fillna('C')

titanic['embarked'].value_counts()
titanic['embarked'] = titanic['embarked'].fillna('S')
titanic['embark_town'] = titanic['embark_town'].fillna('Southampton')

titanic.isnull().sum() # 2)점검

# 3)데이터 기본정보 탐색
titanic.info()

# 4)
titanic.survived.value_counts()
```

```python
# 3-3. 타이타닉 생존율 분석3 - 시각화1
import matplotlib.pyplot as plt

f, ax = plt.subplots(1,2, figsize=(10,5))
titanic['survived'][titanic['sex']=='male'].value_counts().plot.pie(explode=[0,0.1], autopct='%1.1f%%', ax=ax[0])
titanic['survived'][titanic['sex']=='female'].value_counts().plot.pie(explode=[0,0.1], autopct='%1.1f%%', ax = ax[1], shadow = True)

ax[0].set_title('Surviced (Male)')
ax[1].set_title('Surviced (Female)')

plt.show()
```

```python
def category_age(x):
    if x<10: return 0
    elif x<20: return 1
    elif x<30: return 2
    elif x<40: return 3
    elif x<50: return 4
    elif x<60: return 5
    elif x<70: return 6
    else: return 7

titanic['age2'] = titanic['age'].apply(category_age)
# titanic['sex'] = titanic['sex'].map({'male':1, 'female':0}) # map안됨
titanic['sex'] = titanic['sex'].replace({'male':1, 'female':0})
titanic['family'] = titanic['sibsp'] + titanic['parch'] + 1
heatmap_data = titanic[['survived', 'sex', 'age2', 'family', 'pclass','fare']]
colormap = plt.cm.RdBu
sns.heatmap(heatmap_data.astype(float).corr(), linewidths = 0.1, vmax
= 1.0, square = True, cmap = colormap, linecolor = 'white', annot = True,
annot_kws = {"size": 10})
plt.show()
```
