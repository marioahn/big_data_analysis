### 영화 추천 프로그램

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('netflix_titles.csv')
data
```

```python
data = data[data['type']=='Movie'] #영화만 추출
```

```python
data = data[['title', 'cast', 'listed_in', 'description']] # 사용할 column만 다시 저장 사실 title, description만 사용
data
```

```python
# 데이터 확인 결측치 없음
data.info()
```

```python
# 데이터 전처리 => description 열의 값들을 영어만 남기도록
import re
data['description'] = data['description'].map(lambda x : re.sub(r'[^ a-zA-z]+', '', x))
data[['description']]
```

```python
data[data['description'].str.contains(r'[^ a-zA-z]+')] # 잘됐는지 확인
```

```python
# description 리스트에 저장
description = []
for i in data['description']:
    description.append(i)
```

```python
description # 확인
```

```python
# 단어 벡터화
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
tfidf.fit(description)
description_tfidf = tfidf.transform(description)
```

```python
# cosine 유사도 활용
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(description_tfidf, description_tfidf)
```

```python
cosine_sim
```

```python
# 영화 제목을 인덱스로, 인덱스 번호를 값으로 변경한 데이터 프레임 새로 생성
indies = pd.DataFrame(data=data.index, columns=['idx'], index=data['title'].values)
indies
```

```python
# 확인
indies.loc['Zozo']
```

```python
def movie_recommend(title):
    Recommend = []
    idx = indies.loc[title]

    # 안에 있는 유사도를 기준으로 내림차순 정렬 // key => key값을 기준으로함 // lambda x : x[1] 2차원 리스트 안의 리스트들 중 1번 인덱스값을 반환 // reverse = True 내림차순하겠다는 뜻~
    i_num = sorted(enumerate(cosine_sim[idx][0]), key=lambda x : x[1], reverse= True)[1:11] #상위 10개 영화
    # key = lambda x: x[1], reverse = True 2차원 리스트 안의 리스트들 중 1번 인덱스값을 기준으로 내림차순을 하겠다는 뜻~
    
    for i in range(len(i_num)):
        Recommend.append(i_num[i][0])
    return Recommend
```

```python
# 검색한 영화와 비슷한 영화 추천 10개
print('검색한 영화와 비슷한 영화추천 목록 10개')
M = movie_recommend(input())
for i, val in enumerate(M):
    print(i+1, ':', data['title'].iloc[val])
```
